```python
from pdfminer.layout import LAParams, LTTextBox
from pdfminer.pdfpage import PDFPage
from pdfminer.pdfinterp import PDFResourceManager
from pdfminer.pdfinterp import PDFPageInterpreter
from pdfminer.converter import PDFPageAggregator
from pdfminer.converter import TextConverter
import io
import glob
import os
```

```python
def pdf_to_text(filename):
    resource_manager = PDFResourceManager()
    fake_file_handle = io.StringIO()
    converter = TextConverter(resource_manager, fake_file_handle, laparams=LAParams())
    page_interpreter = PDFPageInterpreter(resource_manager, converter)

    with open(filename, 'rb') as fh:

        for page in PDFPage.get_pages(fh,
                                      caching=True,
                                      check_extractable=False): # we changed this from True for this special RCGP pdf 2019-0378-Royal-College-of-General-Practitioners.pdf
            page_interpreter.process_page(page)

        text = fake_file_handle.getvalue()

    # close open handles
    converter.close()
    fake_file_handle.close()
  
    # save output as a txt file
    output = filename + '.txt'
    with open(output, "w") as text_file:
        text_file.write(text)
```

```python
path = 'downloads'
for filename in glob.glob(os.path.join(path, '*.pdf')):
    output = filename + '.txt'
    if not os.path.isfile(output):
        print('processing {}\n'.format(filename))
        try:
            pdf_to_text(filename)
        except Exception as inst:
            print(inst)
            print('Encountered unknown error. Continuing.')
        
        
```
