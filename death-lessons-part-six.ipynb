```python
import pickle
import pandas as pd

# open the pickle we saved earlier of the scraped summary from the death urls

with open('people_data.pickle', 'rb') as handle:
    p = pickle.load(handle)
          
```

```python
# make a dataframe

df = pd.DataFrame(p.keys())
df.columns = ['person']
df.index = df.person
```

```python
# populate our dataframe with the six data fields that most report summaries include...

for person in p.keys():
    mask = person
    df.loc[mask, 'date_of_report'] = p[person][0] # date of report
    df.loc[mask, 'ref'] = p[person][1] # ref
    df.loc[mask, 'name_of_deceased'] = p[person][2] # name of deceased
    df.loc[mask, 'coroner_name'] = p[person][3] # coroner name
    df.loc[mask, 'coroner_area'] = p[person][4] # coroner area
    df.loc[mask, 'category'] = p[person][5] # category
```

```python
df['category'] = df['category'].str.upper()
```

```python
df['category'] = df['category'].str.replace('<P>','')
df['category'] = df['category'].str.replace('</P>','')
```

```python
# these correspond to content free pages e.g https://www.judiciary.uk/publications/railwayrelateddeaths/
df[df['category'].str.contains('<P CLASS=')].person.values 
```

```python
# should be handled better but for now...
# discard them
df = df[~df['category'].str.contains('<P CLASS=')]
```

```python
# some report summaries don't include a reference
# e.g https://www.judiciary.uk/publications/margaret-spencer/
df[df['category'].str.contains('<A')].person.values 
```

```python
# should be handled better but for now...
# discard them
df = df[~df['category'].str.contains('<A')]
```

```python
# some report summaries don't include a coroner name
# e.g https://www.judiciary.uk/publications/omarian-brooks/
# some report summaries don't include a reference
# e.g https://www.judiciary.uk/publications/jamil-ahmed/
df[df.category.str.contains('THIS REPORT IS BEING SENT TO')]
```

```python
# should be handled better but for now...
# discard them
df = df[~df.category.str.contains('THIS REPORT IS BEING SENT TO')]
```
