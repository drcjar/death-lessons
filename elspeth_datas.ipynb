```python
df3['name_stripped'] = df3['name'].astype(str).str.strip()
```

```python
df = df_deduplicated.copy()
```

```python
df['Surname'] = df['name'].astype(str).str.strip().str.split().str[-1]
```

```python
df.astype(str).str.strip().str.split().str[-1]
```

```python
df['Forename'] = df['name'].astype(str).str.strip().str.split().apply(lambda parts: ', '.join(parts[:-1]))
```

```python
# Generate 7-digit unique numbers and prefix with "687" to form a 10-digit ID
prefix = "687"
num_required = len(df)

# Ensure no collisions
np.random.seed(0)
suffixes = np.random.choice(range(1000000, 9999999), size=num_required, replace=False)
df['Unique Identifier'] = [prefix + str(suffix).zfill(7) for suffix in suffixes]
```

```python
df[['Unique Identifier','Forename', 'Surname']].to_csv('pfd_onno_times.csv', index=False)
```
